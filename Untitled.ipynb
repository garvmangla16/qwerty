```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
bc = pd.read_csv('/Users/garvmangla/Desktop/breastcancerdatabase.csv', header=None)
bc.columns=['Age','Race','Marital Status','T Stage','N Stage','6th Stage','Differentiate','Grade','A Stage','Tumor Size','Estrogen Status','Progesterone Status','Regional Node Examined','Reginol Node Positive','Survival Months','Status']
#print(bc.dtypes)
```

```python
#for i in (bc.columns):
       #print(bc[i].unique())
#anomaly detected only in 'Grade'
a=bc.loc[bc['Grade']== 'anaplastic; Grade IV']
print("Anomalies detected =",len(a))
```

```python
#removing the impure data
bc_filtered = bc.loc[bc['Grade']!= 'anaplastic; Grade IV']
print(len(bc_filtered))
#Checking if the impure data is removed
print(bc_filtered['Grade'].unique())
db=bc_filtered.drop('Status', axis = 1).copy()
spc=bc_filtered['Status'].copy()
```

```python
#Converting the data type in 'Grade' from object to int
convert_= { 'Grade': int}
db=db.astype(convert_)
#print(db.dtypes)
```

```python
#Doing one hot encoding
db_encoded=pd.get_dummies(db, columns=["N Stage",
                              'Grade',
                              'Differentiate',
                              '6th Stage',
                              'T Stage',
                              'Marital Status',
                              'Race',
                              'A Stage',
                              'Estrogen Status',
                              'Progesterone Status'])
#print(db_encoded.head())
#finished encoding the database
```

```python
#Now we are ready to make a decision tree but first we make a preliminary decision tree
db_train, db_test, spc_train, spc_test = train_test_split(db_encoded, spc, random_state = 11, train_size= 0.70)
clf_dt= DecisionTreeClassifier()
clf_dt=clf_dt.fit(db_train, spc_train)
#plt.figure(figsize=(400,100))
```

```python
a = plot_tree(clf_dt,
          filled=True,
          rounded=True,
          class_names=["Alive","Dead"],
          feature_names= db_encoded.columns)
```

```python
ConfusionMatrixDisplay.from_estimator(
    clf_dt, db_test, spc_test)
```

```python
path = clf_dt.cost_complexity_pruning_path(db_train, spc_train)
alphas=path.ccp_alphas#################
alphas=alphas[:-1]#####################
clf_dts = [] #The new decision trees will be stored in this array
#Now creating a decision tree for each value of alpha
for i in alphas:
    clf_dt = DecisionTreeClassifier(ccp_alpha=i)
    clf_dt.fit(db_train, spc_train)
    clf_dts.append(clf_dt)
```

```python
train_scores = [clf_dt.score(db_train, spc_train) for clf_dt in clf_dts ]
test_scores = [clf_dt.score(db_test, spc_test) for clf_dt in clf_dts ]

fig, ax = plt. subplots()
ax.set_xlabel("alpha")
ax. set_ylabel("accuracy")
ax.set_title("Accuracy vs alpha for training and testing sets")
ax. plot(alphas, train_scores,marker='o', label="train", drawstyle="steps-post")
ax.plot(alphas, test_scores, marker='o', label="test", drawstyle="steps-post")
ax. legend ()
plt. show()
```

```python
# we can see that the values of alpha between 0.001 and 0.002 give nearly same accuracy....so we can choose any value in that range
```

```python
clf_dt = DecisionTreeClassifier(random_state=6, ccp_alpha=0.00017) # create the tree with ccp_alpha=0.016
## now use 5-fold cross validation create 5 different training and testing datasets that are then used to train and test the tree.
scores = cross_val_score(clf_dt, db_train, spc_train, cv=10)
df = pd.DataFrame(data={'tree': range(10), 'accuracy': scores})
df.plot(x='tree', y='accuracy', marker='o', linestyle='dashdot')
```

```python
# We can not rely on the value of alpha that we got through accuracy vs alpha 
# coz that is for a particular distribution of data into training and testing sets.
# So we need to cross validate the value of alpha for mulitple such distributions and then take its appropriate value.

# We have to create an array to store the result of each fold during cross validation 
alphas_loop_values = []
for i in alphas:
    clf_dt = DecisionTreeClassifier( random_state= 11, ccp_alpha=i)
    scores= cross_val_score(clf_dt, db_train, spc_train, cv=5)
    alphas_loop_values.append([i, np.mean(scores), np.std(scores)])

#Now draw a graph for meanvalues and standard deviations of the scores

alpha_results=pd.DataFrame(alphas_loop_values, columns= ['alpha', 'mean', 'std dev'])
alpha_results.plot(x='alpha', y='mean', yerr='std dev', marker='o', linestyle ='solid', figsize=(75,50))
```

```python
alpha_results [(alpha_results['mean'] > 0.90)]
```

```python
# We take that value of alhpha for which mean has highest value
# First we locate the index for which mean is highest
mean_max_index = alpha_results['mean'].idxmax()
print(mean_max_index)
# Now we extract the value of alpha for that index an it is required alpha
ideal_cpp_alpha = alpha_results.iat[mean_max_index,0]
print(ideal_cpp_alpha)
```

```python
#Now with the desired value of alpha we draw the pruned classification tree
clf_dt_pruned= DecisionTreeClassifier(random_state=0, ccp_alpha=0.0019441)
clf_dt_pruned=clf_dt_pruned.fit(db_train, spc_train)
#plt.figure(figsize=(400,150))
a = plot_tree(clf_dt_pruned,
          filled=True,
          rounded=True,
          class_names=["Alive","Dead"],
          feature_names= db_encoded.columns)

# Now plotting the confusion matrix
ConfusionMatrixDisplay.from_estimator(clf_dt_pruned,db_test, spc_test)
```
